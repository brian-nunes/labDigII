Configuração do cliente MQTT

```python
import paho.mqtt.client as mqtt
import time

broker = "3.141.193.238"            
port = 80                           
keepAlive = 60                      
topico = user+"/slider"                 

def on_connect(client, userdata, flags, rc):
    print("Conectado com codigo " + str(rc))
    client.subscribe(topico, qos=0)

def on_message(client, userdata, msg):
    mensagem = msg.payload.decode("utf-8")
    print("slider:"+mensagem, flush=True)

client = mqtt.Client()                      
client.on_connect = on_connect              
client.on_message = on_message              
client.username_pw_set(user, passwd)        
client.connect(broker, port, keepAlive)
```

Coloca o servo na posição 00

```python
client.loop_start()
client.publish(user+"/E1", payload="0", qos=0, retain=False)
client.publish(user+"/E2", payload="0", qos=0, retain=False)
time.sleep(1)
client.loop_stop()
```

Coloca o servo na posição 01

```python
client.loop_start()
client.publish(user+"/E1", payload="1", qos=0, retain=False)
client.publish(user+"/E2", payload="0", qos=0, retain=False)
time.sleep(1)
client.loop_stop()
```

Coloca o servo na posição 10

```python
client.loop_start()
client.publish(user+"/E1", payload="0", qos=0, retain=False)
client.publish(user+"/E2", payload="1", qos=0, retain=False)
time.sleep(1)
client.loop_stop()
```

Coloca o servo na posição 11

```python
client.loop_start()
client.publish(user+"/E1", payload="1", qos=0, retain=False)
client.publish(user+"/E2", payload="1", qos=0, retain=False)
time.sleep(1)
client.loop_stop()
```

Loop infinito que fica alterando as quatro posições do servo motor

```python
client.loop_start()
while(True):
  client.publish(user+"/E1", payload="0", qos=0, retain=False)
  client.publish(user+"/E2", payload="0", qos=0, retain=False)
  time.sleep(2)

  client.publish(user+"/E1", payload="1", qos=0, retain=False)
  client.publish(user+"/E2", payload="0", qos=0, retain=False)
  time.sleep(2)

  client.publish(user+"/E1", payload="0", qos=0, retain=False)
  client.publish(user+"/E2", payload="1", qos=0, retain=False)
  time.sleep(2)

  client.publish(user+"/E1", payload="1", qos=0, retain=False)
  client.publish(user+"/E2", payload="1", qos=0, retain=False)
  time.sleep(2)
client.loop_stop()
```
